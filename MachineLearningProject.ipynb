```python
import pandas as pd
import numpy as np

trainData = pd.read_csv('train.csv')
```

```python
trainData
```

```python
#replace all NaN values with an empty space
trainData = trainData.fillna(' ')
```

```python
#pad all text entries out to 280 chars, the max limit for twitter
for i, row in trainData.iterrows():
    newVal = row['text'].ljust(280, ' ')
    trainData.at[i,'text'] = newVal
```

```python
#split out each char into its own column
features = trainData.text.str.split('', expand=True)
#convert each char to its ASCII value
for i, row in features.iterrows():
  for j in range(0, 282):
    if(features.at[i,j] == ""):
      features.at[i,j] = ord(" ")
      continue
    features.at[i, j] = ord(features.at[i,j])
#ensure data is structured properly for the model
features = np.asarray(features).astype('float32')
features
```

```python
from tensorflow import keras
from tensorflow.keras import layers
#define model layers, this is just an initial guess
model = keras.Sequential(
    [
        layers.Dense(280, activation="relu", name="layer1"),
        keras.layers.Dropout(.6),
        layers.Dense(140, activation="relu", name="layer2"),
        layers.Dense(2, activation="relu", name="layer6")
    ]
)
```

```python
from keras import losses
#train the model and track results
target = trainData['target']
#model.compile(optimizer="Adam", loss="binary_crossentropy", metrics=["accuracy"])
model.compile(optimizer="Adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.fit(features, target, epochs=100)
```

```python
#results: text-only, padding, 4 hidden layers, 43% accuracy
#results: text-only, padding, 1 hidden layer, 57% accuracy
```

```python
import tensorflow as tf
trainData = pd.read_csv('train.csv')

#replace all NaN values with an empty space
trainData = trainData.fillna(' ')

for i, row in trainData.iterrows():
  if row['text'] == '':
    row['text'] = ' '

#define model layers, instead of splitting text into chars and using ascii values this model will use text vectorization
textLayer = layers.TextVectorization(max_tokens=280, output_mode='int', output_sequence_length=140, name="textVect")
textLayer.adapt(tf.convert_to_tensor(trainData['text']))
model2 = keras.Sequential(
    [
        textLayer,
        #layers.Dense(280, activation="relu", name="layer1"),
        #keras.layers.Dropout(.2),
        layers.Dense(140, activation="relu", name="layer2"),
        layers.Dense(2, activation="relu", name="layer6")
    ]
)
model2.compile(optimizer="Adam", loss="binary_crossentropy", metrics=["accuracy"])
model2.fit(trainData['text'], trainData['target'], epochs=100)
```
